## ChatGPT Webscraping

### Start

````python
import requests
from bs4 import BeautifulSoup
````

### chat_gpt prompt

Website: 'https://subslikescript.com/movies'

```bash
Prompt messages
Scrape this website https://subslikescript.com/movies with python and beautiful soup
locate the element with tag "ul" and class "scripts-list". Scrape all the "a" elements inside.
Get the text attributes and print it.
```

````python
url = 'https://subslikescript.com/movies'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

if response.status_code == 200:
# Step 2: Parse the HTML
    soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Locate the 'ul' with class 'scripts-list'
scripts_list = soup.find('ul', class_='scripts-list')
if scripts_list:
    # Step 4: Find all 'a' elements inside the list
    links = scripts_list.find_all('a')
    # Step 5: Extract and print the text from each 'a' tag
    for link in links:
        print(link.text.strip())
    else:
        print("No ul element with class 'scripts-list' found.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

````
